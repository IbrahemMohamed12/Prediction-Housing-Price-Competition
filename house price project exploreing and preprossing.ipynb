### Data Preprocessing

```python
# because it's in data description categorical data( MSSubClass: Identifies the type of dwelling involved in the sale.)
data1['MSSubClass'] = data1['MSSubClass'].astype(str)
```

### Fill categorical values

```python
data2 = data1.copy()
```

```python
# Select columns with datatype 'object' from the DataFrame data2
Object_data = data2.select_dtypes(include='object')

# Calculate the count of missing values for each column in Object_data
missing_obj = Object_data.isna().sum()

# Filter to include only columns with missing values (count > 0)
missing_obj[missing_obj > 0]
```

```python
# Impute Using a constant value ('None') for specific columns
for column in [
    'Alley',
    'BsmtQual',
    'BsmtCond',
    'BsmtExposure',
    'BsmtFinType1',
    'BsmtFinType2',
    'FireplaceQu',
    'GarageType',
    'GarageFinish',
    'GarageQual',
    'GarageCond',
    'PoolQC',
    'Fence',
    'MiscFeature'
]:
    # Check if the column exists in the DataFrame
    if column in data2.columns:
        # Fill missing values with 'None' for the current column
        data2[column] = data2[column].fillna('None')
    else:
        print(f"Column '{column}' not found in the DataFrame.")

```

```python
# Impute missing values using the column mode for specific columns
for column in [
    'MSZoning',
    'Utilities',
    'Exterior1st',
    'Exterior2nd',
    'MasVnrType', 
    'Electrical', 
    'KitchenQual',
    'Functional',
    'SaleType'
]:
    # Check if the column exists in the DataFrame
    if column in data2.columns:
        # Fill missing values with the mode of the current column
        data2[column] = data2[column].fillna(data2[column].mode()[0])
    else: 
        # Print a warning if the column is not found in the DataFrame
        print(f"Column '{column}' not found in the DataFrame.")
```

### fill numarecal values

```python
# Select columns with datatype 'number' from the DataFrame data2
Object_data = data2.select_dtypes(include='number')

# Calculate the count of missing values for each column in num_data
missing_obj = Object_data.isna().sum()

# Filter to include only columns with missing values (count > 0)
missing_obj[missing_obj > 0]
```

```python
def Knn_impute(df, na_target):
    df = df.copy()

    numeric_df = data2.select_dtypes('number')    
    
    non_na_columns = numeric_df.loc[:, numeric_df.isna().sum() == 0].columns
    
    Y_train = numeric_df.loc[numeric_df[na_target].isna() == False, na_target]
    X_train = numeric_df.loc[numeric_df[na_target].isna() == False, non_na_columns]
    X_test  = numeric_df.loc[numeric_df[na_target].isna() == True, non_na_columns]
    
    Knn = KNeighborsRegressor()
    Knn.fit(X_train, Y_train)

    y_pred = Knn.predict(X_test)
    
    df.loc[df[na_target].isna() == True, na_target] = y_pred
    
    return df
```

```python
for column in [
    'LotFrontage',
    'MasVnrArea',
    'BsmtFinSF1',
    'BsmtFinSF2',
    'BsmtUnfSF',
    'TotalBsmtSF',
    'BsmtFullBath',
    'BsmtHalfBath',
    'GarageYrBlt',
    'GarageCars',
    'GarageArea'
]:
    data2 = Knn_impute(data2, column)
```
